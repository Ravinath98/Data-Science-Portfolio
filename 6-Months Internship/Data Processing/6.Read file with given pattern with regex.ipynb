```python
#import libraries...
import pandas as pd
import glob
import re
```

```python
#here I rename 2 files inside the folder forcast_july,  july3.xls --> july@3.xls & july4.xls --> july@4.xls to differentiate the pattern with other files inside that folder
path_folder=r"E:\Desktop\Datasets\forcast_july"
filenames_xls = glob.glob(path_folder + "\*.xls") #xls extensions..
#print(filenames_xls)
pattern="()[a-z]+@[0-9]+\.xls" #here in the begining it's matches anything and at the end it stick with the pattern I given here
for file in filenames_xls: #iterate over files
    #print(file)
    if re.search(pattern,file): #check,if matches the pattern by each file(there are 2 matches files)
        print(file)
        df=pd.read_excel(file)
        print(df.head())
        print('-------------------------------------------------------------------------------------------------') 
```

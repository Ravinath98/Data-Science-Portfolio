It can be seen that, there are outliers in the price columns.So,let's remove them

```python
q1,q3=np.percentile(df['price'],[25,75])
```

```python
q1
```

```python
q3
```

```python
iqr=q3-q1
iqr
```

```python
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
```

```python
def findoutliersIQR(data):
    outliersIQR=[]
    for i in data:
        if i>upper_bound or i<lower_bound:
            outliersIQR.append(i)
    return outliersIQR  
```

```python
out=findoutliersIQR(df['price'])
len(out)
```

there are 240 outliers present in the dataset.1st lets remove them and make the evaluation.Then again check by,replacing them with mean

```python
#one hot encoding...
df3=pd.get_dummies(df2)
```

```python
df3.columns
```

## Define x and y

```python
x=df3.drop('price',axis=1)
y=df3['price']
```

## Train test split

```python
#import the library
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

## Standardization the dataset

```python
from sklearn.preprocessing import StandardScaler #import the required library
scaling=StandardScaler() #make the scaling object
x_train=scaling.fit_transform(x_train) #scaling all the time columns dispatch values...
x_test=scaling.fit_transform(x_test)
```

## Apply different regression models & do the evaluation

```python
#evaluation libraries
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
# Linear Regression
lr = LinearRegression()
lr.fit(x_train, y_train)
predictions = lr.predict(x_test)
r2= r2_score(predictions, y_test)
mae=mean_absolute_error(predictions, y_test)
mse=mean_squared_error(predictions, y_test)
print("MAE:",mae)
print("MSE:",mse)
print("R2 Score:", r2)
```

```python
# support vector regression
svr = SVR(C=100000)
#svr = SVR()
svr.fit(x_train, y_train)
predictions = svr.predict(x_test)
r2 = r2_score(predictions, y_test)
mae=mean_absolute_error(predictions, y_test)
mse=mean_squared_error(predictions, y_test)
print("MAE:",mae)
print("MSE:",mse)
print("R2 Score:", r2)
```

```python
#RandomForestRegressor
rf = RandomForestRegressor(n_estimators=50)
rf.fit(x_train, y_train)
predictions = rf.predict(x_test)
r2 = r2_score(predictions, y_test)
mae=mean_absolute_error(predictions, y_test)
mse=mean_squared_error(predictions, y_test)
print("MAE:",mae)
print("MSE:",mse)
print("R2 Score:", r2)
```

```python
#Lasso regression
lar = Lasso()
lar.fit(x_train, y_train)
predictions = lar.predict(x_test)
r2= r2_score(predictions, y_test)
mae=mean_absolute_error(predictions, y_test)
mse=mean_squared_error(predictions, y_test)
print("MAE:",mae)
print("MSE:",mse)
print("R2 Score:", r2)
```

```python
#Ridge regression
rr = Ridge()
rr.fit(x_train, y_train)
predictions = rr.predict(x_test)
r2= r2_score(predictions, y_test)
mae=mean_absolute_error(predictions, y_test)
mse=mean_squared_error(predictions, y_test)
print("MAE:",mae)
print("MSE:",mse)
print("R2 Score:", r2)
```

```python
#XGBRegressor
xfb = XGBRegressor()
xfb.fit(x_train, y_train)
predictions = xfb.predict(x_test)
r2= r2_score(predictions, y_test)
mae=mean_absolute_error(predictions, y_test)
mse=mean_squared_error(predictions, y_test)
print("MAE:",mae)
print("MSE:",mse)
print("R2 Score:", r2)
```

```python
#ElasticNet
en = ElasticNet()
en.fit(x_train, y_train)
predictions = en.predict(x_test)
r2= r2_score(predictions, y_test)
mae=mean_absolute_error(predictions, y_test)
mse=mean_squared_error(predictions, y_test)
print("MAE:",mae)
print("MSE:",mse)
print("R2 Score:", r2)
```

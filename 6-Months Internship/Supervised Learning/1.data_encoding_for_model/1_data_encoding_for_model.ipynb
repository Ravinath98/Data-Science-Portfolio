```python
#import pandas
import pandas as pd
```

Ordinal Encoding(Label Encoding)

```python
import category_encoders as ce #import the library for encoding categorical variables into numeric 
```

```python
#read the dataset
df1=pd.read_csv('healthy_lifestyle_city_2021.csv')
```

```python
#head values(1st 5)
df1.head()
```

```python
# creating the object of the OrdinalEncoding
OE = ce.OrdinalEncoder(cols=['City'])# fit and transform to get the encoded data
df1['Transformed_City_OE']=OE.fit_transform(df1['City'])
```

```python
#after transforming....
df1.head(20)
```

One Hot encoding

```python
# create an object of the OneHotEncoder
OHE = ce.OneHotEncoder(cols=['City'])# fit and transform to get the encoded data
df2=OHE.fit_transform(df1['City'])
```

```python
#after transforming
df3=pd.concat([df1, df2], axis=1)
df4=df3.drop('Transformed_City_OE',axis=1)
df4.head(20)
```

Binary encoding

```python
# 1)categorical to numeric(using ordinal encoder) 2)numberic to binary 3)binary value is split into different columns
# create an object of the BinaryEncoder
BE = ce.BinaryEncoder(cols=['City'])# fit and transform to get the encoded data
df5=BE.fit_transform(df1['City'])
```

```python
#after transforming
df6=pd.concat([df1, df2], axis=1)
df7=df6.drop('Transformed_City_OE',axis=1)
df7.head(20)
```

BaseN encoding

```python
# create an object of the BaseNEncoder
BN = ce.BaseNEncoder(cols=['City'],base=3)# fit and transform to get the encoded data #in base 3
df8=BN.fit_transform(df1['City'])
```

```python
#after transforming
df9=pd.concat([df1, df8], axis=1)
df10=df9.drop('Transformed_City_OE',axis=1)
df10.head(20)
```

Hashing 

```python
# create an object of the Hashing Encoder
HE = ce.HashingEncoder(cols=['City'],n_components=8)# fit and transform to get the encoded data #8 bins..
df11=HE.fit_transform(df1['City'])
```

```python
#after transforming
df12=pd.concat([df1, df11], axis=1)
df13=df12.drop('Transformed_City_OE',axis=1)
df13.head(20)
```
